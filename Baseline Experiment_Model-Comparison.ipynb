```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# =========================
# 1. ENTER YOUR RESULTS
# =========================

# Leaf-level results
leaf_results = pd.DataFrame({
    "Model": ["RandomForest", "SVM", "LogisticRegression", "GradientBoosting"],
    "Accuracy": [0.474569, 0.502980, 0.491267, 0.497878],
    "F1": [0.431970, 0.391984, 0.416395, 0.442235]
})

# Tree-level results
tree_results = pd.DataFrame({
    "Model": ["RandomForest", "SVM", "LogisticRegression", "GradientBoosting"],
    "Accuracy": [0.596154, 0.576923, 0.519231, 0.557692],
    "F1": [0.511628, 0.388889, 0.528302, 0.566038]
})

# =========================
# 2. MERGE RESULTS
# =========================

# Merge leaf and tree results
merged = leaf_results.merge(tree_results, on="Model", suffixes=("_Leaf", "_Tree"))

# =========================
# 3. PLOT SETTINGS
# =========================

models = merged["Model"]
x = np.arange(len(models))
width = 0.35

# =========================
# 4. CREATE FIGURE (ACCURACY)
# =========================

plt.figure(figsize=(8, 5))

plt.bar(x - width/2, merged["Accuracy_Leaf"], width, label="Leaf-level")
plt.bar(x + width/2, merged["Accuracy_Tree"], width, label="Tree-level")

plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.title("Model Comparison: Leaf-level vs Tree-level Classification")
plt.xticks(x, models, rotation=20)
plt.legend()

plt.tight_layout()
plt.savefig("Fig4_Model_Comparison_Accuracy.png", dpi=300)

# Save as PDF
plt.savefig("Model_Comparison_Accuracy.pdf", format="pdf")

# Save as EPS (optional)
plt.savefig("Model_Comparison_Accuracy.eps", format="eps")

plt.show()

# =========================
# 5. CREATE FIGURE (F1-SCORE)
# =========================

plt.figure(figsize=(8, 5))

plt.bar(x - width/2, merged["F1_Leaf"], width, label="Leaf-level")
plt.bar(x + width/2, merged["F1_Tree"], width, label="Tree-level")

plt.xlabel("Model")
plt.ylabel("F1-score")
plt.title("Model Comparison: Leaf-level vs Tree-level Classification (F1-score)")
plt.xticks(x, models, rotation=20)
plt.legend()

plt.tight_layout()
plt.savefig("Fig4_Model_Comparison_F1.png", dpi=300)

# Save as PDF
plt.savefig("Model_Comparison_F1.pdf", format="pdf")

# Save as EPS (optional)
plt.savefig("Model_Comparison_F1.eps", format="eps")

plt.show()
```
